# Overview

The motion picture industry is a multi-billion dollar business with movie success being largely uncertain, with “hits” and “flops” released almost every year. Analytics about the box-office performance of movies can give useful information to Microsoft so that they can make the most strategic decisions regarding production and financing. The analysis led to three key findings and recommendations for Microsoft:
1. Microsoft should invest in animated films, which offer the largest risk/reward equation in the movie business
2. Acquiring or partnering with a studio could be a viable strategy for Microsoft as the industry continues to consolidate
3. When box office profits and generating the quickest return on investment are important, then the months of summer are the ideal time of the year for Microsoft to release a film

## Business Problem

Microsoft (<i>the "Client"</i>) is investigating creating a new movie studio, but they don’t know anything about creating movies. In this report, we will explore what types of films have been most successful at the box office over the past 10 years. 

It's important to note upfront that while there's a lot of money to be made in the film industry, the economics of movie-making are far from simple.
There's no sure path for a film to turn a profit since factors like brand awareness, P&A budgets, and the desires of a fickle public come into play.
Theater attendance in the U.S. has been challenging over recent years, making it even more important to earn money in foreign theaters.
Ever since Star Wars, merchandising has played a major role in revenue for films that appeal to children.
Television rights, video-on-demand, and streaming services are increasingly important sources of income for movie studios.

In this analysis, I will attempt to resolve three key questions for the Client in an effort to make a transition into movie-making successful:
  1. What types of movies have been produced the most in the past 20 years and what genres have been most successful?
  2. Which studios have been the most successful and do any warrant a potential partnership?
  3. Is release date a determinant of box office success?

### Most Produced Movies by Genre

```python
genres_count = genres['genres'].value_counts()
genres_count = genres_count[:10,]

plt.figure(figsize=(15, 7))
sns.set_context('notebook')
sns.set_style("white")
sns.barplot(
    x=genres_count.index,
    y=genres_count.values,
    palette="cubehelix"
).set(
    ylabel="Number of Movies",
    xlabel="Top 10 Genres",
    title="Most Produced Genres"
    
);

#remove top and right axes spines
sns.despine()
```

### Most Profitable Genres

```python
#Join budgets and genres on shared 'primary_title' column
genre_num = pd.merge(genres, budgets, on='primary_title', how='right')
genre_num.info()
genre_num.head()
```

```python
#Drop columns for readability
genre_num.drop(['tconst', 'original_title', 'start_year', 'runtime_minutes', 'averagerating', 'numvotes', 'id'], axis=1, inplace=True)
```

```python
genre_num.head()
```

```python
#Calculate gross profit and create new column for it
genre_num['gross_profit'] = genre_num['worldwide_gross'] - genre_num['production_budget']
genre_num.head()
```

```python
#Find null values
display(genre_num.isna().sum())
```

```python
#Drop na values in genres
genre_num.dropna(subset=["genres"], inplace=True)
```

```python
#Check to see if values were actually dropped
display(genre_num.isna().sum())
```

```python
genre_group = genre_num.groupby(genre_num.genres).sum()
genre_group.head(20)
```

```python
#Taking the top 10 genres and sorting by the average gross profit
genre_num.groupby('genres').mean()['gross_profit'].iloc[:10,].sort_values(ascending=False)
```

```python
order= genre_num.groupby('genres').mean()['gross_profit'].iloc[:10,].sort_values(ascending=False).index

from matplotlib.ticker import FuncFormatter

def millions(x, pos):
    #The two args are the value and tick position'
    return '$%1.0fM' % (x * 1e-6)

formatter = FuncFormatter(millions)

fig, ax = plt.subplots(figsize=(15,7),)
sns.set_context('notebook')
sns.set_style("white")
ax.yaxis.set_major_formatter(formatter)
ax = sns.barplot(data=genre_num, x='genres',y='gross_profit',order=order
).set(
ylabel="Gross Profit (millions)",
xlabel="Movie Genres", 
title="Most Profitable Genres"
)
;
sns.despine()
```

### Top Studios

```python
top_stud = studios.groupby(['studio']).sum().sort_values(by='total_gross', ascending=False)
top_stud = top_stud.iloc[:10,]
top_stud
```

```python
def billions(x, pos):
    #The two args are the value and tick position'
    return '$%1.0fB' % (x * 1e-9)

formatter = FuncFormatter(billions)

fig, ax = plt.subplots(figsize=(15,7),)
ax.yaxis.set_major_formatter(formatter)
ax.tick_params(axis='x', labelrotation = 45)
sns.barplot(
    x=top_stud.index,
    y='total_gross', 
    data=top_stud, 
    
).set(
    ylabel="Gross Sales",
    xlabel="Studios",
    title="Highest Grossing Studios"
);

sns.despine()

#plt.savefig('Highest_Grossing_Studios.png', bbox_inches = 'tight')
```

### Release Date


```python
#Determining average revenue per month and visualizing it
month_average = budgets[['release_month', 'worldwide_gross']]
month_average_grouped = month_average.groupby('release_month', as_index=False).mean()
month_average_grouped
```

```python
#Converting to months for readability 
months=['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
month_average_grouped['release_month'] = months
month_average_grouped
```

```python
import matplotlib.dates as mdates
from matplotlib.dates import DateFormatter

def millions (x, pos):
    #The two args are the value and tick position'
    return '$%1.0fM' % (x * 1e-6)

formatter = FuncFormatter(millions)


fig, ax = plt.subplots(figsize=(15,7),)
ax.yaxis.set_major_formatter(formatter)
sns.lineplot(
    x='release_month', 
    y='worldwide_gross', 
    data=month_average_grouped,

).set(
    ylabel="Gross Sales",
    xlabel="Months",
    title="Average Box Office Performance by Month"

);

#remove top and right axes spines
sns.despine()
```

# Findings

***

### Animation is the Most Profitable Movie Genre

In terms of genres, animation has the highest average profit, followed by adventure and fantasy. However, animation also has the largest variance in terms of gross profit as shown by the error bar, which suggests it could be a riskier genre to produce. 

![Most Profitable Genres](./images/Most_Profitable_Genres.png)


### Disney and Warner Bros. Lead The Way at the Box Office

![Highest Grossing Studios](./images/Highest_Grossing_Studios.png)


Walt Disney Studios ranks number 1 on this list of the top film production companies with total box office grossing exceeding US$40,000,000,000. The Walt Disney Studios division has several prominent film production companies, including, Walt Disney Pictures, Walt Disney Animation Studios, Pixar, Marvel Studios, Lucasfilm, 20th Century Studios and Searchlight Pictures.

<i>*The acquisition of 21st Century Fox by Disney was held from December 14, 2017 to March 20, 2019. Similar to other Disney film units, distribution of 20th Century films is now handled by Walt Disney Studios Motion Pictures, while 20th Century's sister company, Searchlight Pictures, operates their own autonomous distribution unit.[8] The first film released by Disney under the studio's new name was The Call of the Wild.</i>

### The Summer Box Office Phenomena Is Real


![Most Profitable Genres](./images/Box_Office_Performance_Month.png)

If studios are looking for pure ticket sales and the quickest return on their initial investment (budget), then the summer months present the best opportunity for a release. With children out of school and parents on vacation, this makes sense. 

November can also be a lucrative time for films, as studios gear up the the upcoming awards season.

# Recommendations

This analysis leads to **three key recommendations**: 

1. Microsoft should invest in animated films, which offer the largest risk/reward equation in the movie business
2. Acquiring or partnering with a studio could be a viable strategy for Microsoft as the industry continues to consolidate
3. When box office profits and generating the quickest return on investment are important, then the months of summer are the ideal time of the year for Microsoft to release a film


### Next Steps
Big box office doesn't always mean big profits. The film industry is complex, and ticket sales alone don't drive revenue. There’s merchandising, VOD, streaming video, and a plethora of other distribution channels that can help studios turn a profit. Further investigation is needed to better understand how such factors affect the profitability of films.
